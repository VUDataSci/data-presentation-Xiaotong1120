# Data Imput

```python
# import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
```

```python
# import data
df = pd.read_csv("./data/raw/games.csv")
```

# Data Clean

```python
# clean data

# drop unrelated columns
cols_names = ["Unnamed: 0", "Team", "Reviews", "Summary", "Number of Reviews", "Times Listed"]

df = df.drop(columns = cols_names)

df.head()
```

```python
# convert "Release Date" column to datetime type

# drop three unreleased games
unreleased_games = df[df['Release Date'] == "releases on TBD"]
df = df.drop(unreleased_games.index)

df["Release Date"] = pd.to_datetime(df["Release Date"])
```

```python
# convert string columns to int
cols_names = ["Plays","Playing","Backlogs","Wishlist"]
for col in cols_names:
    df[col] = df[col].str.replace("K","000")
    df[col] = df[col].map(lambda x:int(float(x) * 1000) if "." in x else int(x))
```

```python
# change columns names to make it more comprehensive
df.columns = map(str.lower, df.columns)
df.rename(columns={"release date":"releaseDate","plays":"totalPlayers","playing":"activePlayers"}, inplace = True)
df.sample(3)
```

```python
df.info()
```

```python
df.to_csv("./data/clean/cleaned_games.csv")
```

# The popularity of different genres

```python
# count the popularity of genres
df_1 = df.copy()

# clean the genre column
df_1["genres"] = df_1["genres"].str.replace("\[|\]|'", "")

# split different genres from the same row
genre_series = df_1['genres'].str.split(",\s*").explode()

#remove empty strings
genre_series = genre_series[genre_series.str.strip().astype(bool)]

#count the occurrences
genre_popularity = genre_series.value_counts()

genre_popularity
```

```python
# set the aesthetic style of the plots
sns.set_style("whitegrid")

# set up the figure and axis
plt.figure(figsize = (14, 10))

# create a color palette
palette = sns.color_palette("coolwarm", len(genre_popularity))

# plot the genre popularity data with the new color palette
sns.barplot(x = genre_popularity.index, y = genre_popularity.values, palette = palette)

# add some aesthetic enhancements
plt.title('Popularity of Game Genres', fontsize = 28)
plt.xlabel('Genre', fontsize = 24)
plt.ylabel('Count', fontsize = 24)
plt.xticks(rotation = 45, ha = 'right', fontsize = 20)
plt.yticks(fontsize = 20)

# add value labels on top of each bar
for index, value in enumerate(genre_popularity.values):
    plt.text(index, value, f'{value}', ha = 'center', va = 'bottom', fontsize = 16)

# despine for a cleaner look
sns.despine(top = True, right = True)

# show the plot
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.savefig('./image/Popularity of Game Genres.png')
plt.show()
```

# Median Average Game Ratings by Genre

```python
df_2 = df.copy()

# convert the string representation of lists in 'genres' into actual lists
df_2['genres'] = df_2['genres'].apply(literal_eval)

# explode the dataframe on the 'genres' column to have one genre per row
exploded_genres_df = df_2.explode('genres')

# now group by 'genres' and calculate the average rating for each genre
average_ratings_by_genre = exploded_genres_df.groupby('genres')['rating'].median().sort_values(ascending=False)

average_ratings_by_genre.reset_index()
```

```python
# set the aesthetic style of the plots
sns.set_style("whitegrid")

# create the bar plot
plt.figure(figsize=(14, 10))
bar_plot = sns.barplot(x = average_ratings_by_genre.index, y = average_ratings_by_genre.values, palette = 'coolwarm')

# set the title and labels
bar_plot.set_title('Median Average Game Ratings by Genre', fontsize = 24)
bar_plot.set_xlabel('Genre', fontsize = 22)
bar_plot.set_ylabel('Median Average Rating', fontsize = 22)

# rotate the x labels to show them more clearly
plt.xticks(rotation = 45, ha = 'right', fontsize = 20)
plt.yticks(fontsize = 20)

# add the values on top of the bars
for index, value in enumerate(average_ratings_by_genre.values):
    plt.text(index, value, f'{value:.2f}', ha = 'center', va = 'bottom', fontsize = 18)

# show the plot
plt.tight_layout()  # Adjust the layout to fit the labels
plt.savefig('./image/Median Average Game Ratings by Genre.png')
plt.show()
```

# Median Average Game Ratings by Release Year

```python
df_3 = df.copy()

# get the year of the date 
df_3["year"] = df_3["releaseDate"].dt.year

# delete the year in the future (the ratings have no meanings for the games that haven't been issued)
df_3 = df_3[df_3["year"].isin([2024,2025]) == False]

# calculate the median
average_ratings_by_year = df_3.groupby("year")["rating"].median()
```

```python
# line plot
plt.figure(figsize = (14, 10))

# plot the average ratings by year with seaborn
sns.lineplot(x = average_ratings_by_year.index, y = average_ratings_by_year.values, marker = 'o', linewidth = 2.5)

# add title and labels with increased font sizes for readability
plt.title('Median Average Game Ratings by Release Year', fontsize = 24)
plt.xlabel('Year', fontsize = 22)
plt.ylabel('Median Average Rating', fontsize = 22)

# rotate the x-axis labels for better visibility
plt.xticks(average_ratings_by_year.index, rotation = 45, ha = "right", fontsize = 16)
plt.yticks(fontsize = 20)

# tighten the layout to ensure everything fits without overlapping
plt.tight_layout()
plt.savefig('./image/Median Average Game Ratings by Release Year.png')

# show the plot
plt.show()
```

# Correlation Matrix between Rating and Players

```python
df_4 = df.copy()

# calculate the correlation matrix to understand the relationship between rating and the player-related columns
correlation_matrix = df[['rating', 'totalPlayers', 'activePlayers', 'backlogs', 'wishlist']].corr()
```

```python
# set the aesthetic style of the plots
sns.set_style("whitegrid")

# customize the font sizes using plt.rcParams
plt.rcParams['axes.titlesize'] = 22  # Title font size
plt.rcParams['axes.labelsize'] = 20  # Axis label font size
plt.rcParams['xtick.labelsize'] = 18 # X-axis tick label size
plt.rcParams['ytick.labelsize'] = 18 # Y-axis tick label size
plt.rcParams['legend.fontsize'] = 16 # Legend font size
plt.rcParams['figure.titlesize'] = 20 # Overall figure title size

# plots with loess smoothing lines
fig, axes = plt.subplots(2, 2, figsize = (14, 10))
fig.suptitle('Relationship between Rating and Player-Related Columns (with Loess Smoothing)', fontsize = 24)

# defining a function to annotate the plots with correlation coefficient
def annotate_plot(ax, x, y, data):
    corr_coef = data[[x, y]].corr().iloc[0,1]
    ax.text(0.05, 0.95, f'Corr: {corr_coef:.2f}', transform = ax.transAxes, 
            fontsize = 16, verticalalignment = 'top', bbox = dict(boxstyle = 'round', facecolor = 'white', alpha = 0.5))

# plot with loess smoothing line for Total Players
sns.regplot(ax = axes[0, 0], data = df_4, x = 'rating', y = 'totalPlayers', scatter_kws = {'alpha':0.5}, line_kws = {"color": "red"}, lowess = True)
annotate_plot(axes[0, 0], 'rating', 'totalPlayers', df_4)
axes[0, 0].set_title('Rating vs Total Players')

# plot with loess smoothing line for Active Players
sns.regplot(ax = axes[0, 1], data = df_4, x = 'rating', y = 'activePlayers', scatter_kws = {'alpha':0.5}, line_kws = {"color": "red"}, lowess = True)
annotate_plot(axes[0, 1], 'rating', 'activePlayers', df_4)
axes[0, 1].set_title('Rating vs Active Players')

# plot with loess smoothing line for Backlogs
sns.regplot(ax = axes[1, 0], data = df_4, x = 'rating', y = 'backlogs', scatter_kws = {'alpha':0.5}, line_kws = {"color": "red"}, lowess = True)
annotate_plot(axes[1, 0], 'rating', 'backlogs', df_4)
axes[1, 0].set_title('Rating vs Backlogs')

# plot with loess smoothing line for Wishlist
sns.regplot(ax = axes[1, 1], data = df_4, x = 'rating', y = 'wishlist', scatter_kws = {'alpha':0.5}, line_kws = {"color": "red"}, lowess = True)
annotate_plot(axes[1, 1], 'rating', 'wishlist', df_4)
axes[1, 1].set_title('Rating vs Wishlist')

plt.tight_layout(rect = [0, 0.03, 1, 0.95]) # Adjust tight_layout to take suptitle into account
plt.savefig('./image/Relationship between Rating and Player-Related Columns (with Loess Smoothing).png')
plt.show()
```
